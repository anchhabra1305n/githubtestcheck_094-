```python
def calculate_entropy(data):
    
    label_column = data[:,-1]
    _, counts = np.unique(label_column,return_counts=True) 
    
    probabilities = counts/counts.sum() 
    
    entropy = sum(probabilities * -np.log2(probabilities))
        
    return entropy
```

```python
def calculate_overall_entropy(data_below, data_above):
    
    n_data_points = len(data_below) + len(data_above)
    
    p_data_below =  len(data_below) / n_data_points
    p_data_above =  len(data_above) / n_data_points
    
    overall_entropy = (p_data_below * calculate_entropy(data_below) + 
                  p_data_above * calculate_entropy(data_above))
    
    return overall_entropy
```

```python
def determine_best_split(data, potential_splits):
    overall_entropy = 9999
    for column_index in potential_splits:
        for value in potential_splits[column_index]:
            data_below, data_above = split_data(data, split_column=column_index, split_value=value)
            current_overall_entropy = calculate_overall_entropy(data_below, data_above)

            if current_overall_entropy <= overall_entropy:
                overall_entropy = current_overall_entropy
                best_split_column = column_index
                best_split_value = value
    
    return best_split_column, best_split_value
```

### Algorithm

### Step-1

```python
def decision_tree_algorithm(df)
# This function takes argument as a dataframe but we know our helper functions are built on numpy 2d array 
# (to make the jobs faster)
# Thats why we need to convert our dataframe to numpy 2d array within this function (first call in the function)
# This function is also a recursive function - which can call itself many times
```

### Step-2

```python
def decision_tree_algorithm(df,counter=0):
# since we need to convert df to numpy 2d array, we use counter argument in function with default value is 0
    
    # data preparation
    if counter == 0:
    # first call of the function - if its a first call then convert the pandas dataframe to numpy
        data = df.values
    else:
    # If its not a first call - then keep the data as pandas dataframe
        data = df
```

### Step-3
The base case

```python
# Now if we look at our tree our base case is data purity function (condition) because if it is true then we 
# classify and stop - we need to stop this recursive function oherwise it will keep running
# Then function enters into the recursive part 

# base case
if check_purity(data):
    classification = classify(data)
    return classification

# recursive part
else:
# If the data is not pure then we enter the recursive part of the function
    counter += 1 # We are icreasing our counter by 1
    
    # helper function - now we run all our helper function as per the process flow
    potential_splits = get_potential_splits(data)
    split_column, split_value = determine_best_split(data,potential_splits)
    data_below, data_above = split_data(data,split_column,split_value)
    
    # instantiate sub-tree - for simplicity we keep only first subtree which classify setosa or versicolor - removing 
    # other data points of virginica
    question = "{} <= {}".format(split_column,split_value)
    Sub_tree = {question: []}
    
    # find answers 
    yes_answer =  decision_tree_algorithm(data_below,counter)
    no_answer =  decision_tree_algorithm(data_above,counter)
    
    Sub_tree[question].append(yes_answer)
    Sub_tree[question].append(no_answer)
    
    return Sub_tree
```

### Q - How decision_tree_algorithm function works?

###  Step-1 takes two arguments - pd dataframe and counter value = 0
counter is used to convert the dataframe to numpy 2d array and it should be done only once within the loop thats why a pre value of 0 is set and then counter is increased in the next steps

### Step-2 The base case is run - if the data is pure - run classify function (label column category)

### Step-3 if the data is not pure -
#### step-1 goes to else condition, now we already have data in numpy 2d array so we set the counter = 1 so that the data preparation step does not run 
#### step-2 Now we run our helper functions 
Potential splits (data dictionary with all the column indices and unique values from rows). Best split column and split value from determine_best_split using entropy function post the split cutoff and split column we get the data above and data below the cutoff
#### step-3 Build the subtree based on the question
Sub_tree is a dictionary with key as question and list as values/answers of the questions from data below (Yes) and data above (No)

 * step-1 If the answer is 'yes' or the data is below - function calls itself recursively (decision_tree_algorithm) and takes two arguments (data_below and counter (now value is 1- which means the data will remain numpy 2d array))
    * step-1a recursive decision_tree_algorithm function runs its first step (counter =1) so data will not get converted to numpy 2d array
    * step-1b The base case is run - if the data is pure - run classify function (label column category).This is same as Step-2 from the main steps
    * step-1c if the data is not pure - it runs the same as Step-3 from the main steps (top of the tree)

* step-2 If the answer is 'no' or the data is above - function calls itself recursively (decision_tree_algorithm) and takes two arguments (data_above and counter (now value is 1- which means the data will remain numpy 2d array))
    * step-1a recursive decision_tree_algorithm function runs its first step (counter =1) so data will not get converted to numpy 2d array
    * step-1b The base case is run - if the data is pure - run classify function (label column category).This is same as Step-2 from the main steps
    * step-1c if the data is not pure - it runs the same as Step-3 from the main steps (top of the tree)
* step-3 It will keep appending the questions and 'yes' and 'no' answers to the sub_tree

### This tree will be a fully grown tree until and unless it classfies each and every value of the data

### Step-4
Combining Step-1 to Step-3

```python
def decision_tree_algorithm(df,counter=0):
    
    if counter == 0:
        data = df.values
    else:
        data = df
        
    # base case
    if check_purity(data):
        classification = classify(data)
        return classification

    # recursive part
    else:
        counter += 1 # If data is not pure, we are icreasing our counter by 1
        print(f"Counter values {counter}")
    
    # helper function
    potential_splits = get_potential_splits(data)
    split_column, split_value = determine_best_split(data,potential_splits)
    data_below, data_above = split_data(data,split_column,split_value)
    
        
    # subtree
    question = "{} <= {}".format(split_column,split_value)
    Sub_tree = {question: []}
    
    # find answers 
    yes_answer =  decision_tree_algorithm(data_below,counter)
    no_answer =  decision_tree_algorithm(data_above,counter)
        
    Sub_tree[question].append(yes_answer)
    Sub_tree[question].append(no_answer)
    
    return Sub_tree    
```

```python
def split_data(data,split_column,split_value): 
    
    split_column_values = data[:,split_column] 
    
    type_of_feature = FEATURE_TYPES[split_column] # here split_column is an index here and we are picking up the type from 
    # the index
    
    if type_of_feature == "continuous":
        data_below = data[split_column_values <= split_value]
        data_above = data[split_column_values > split_value]
    
    else: # this is for categorical
        data_below = data[split_column_values == split_value]
        data_above = data[split_column_values != split_value]
        
    return data_below, data_above
```

```python
def decision_tree_algorithm(df,counter=0, min_samples = 3,max_depth = 5):
    
    if counter == 0:
        global COLUMN_HEADERS, FEATURE_TYPES # Passing Feature types as a global variable
        COLUMN_HEADERS = df.columns
        FEATURE_TYPES = determine_type_of_feature(df)
        data = df.values
    else:
        data = df
        
    # base case
    if (check_purity(data)) or (len(data) < min_samples) or (counter == max_depth): # here added one condition 
        classification = classify(data)
        return classification

    # recursive part
    else:
        counter += 1 # If data is not pure, we are icreasing our counter by 1
    
        # helper function
        potential_splits = get_potential_splits(data)
        split_column, split_value = determine_best_split(data,potential_splits)
        data_below, data_above = split_data(data,split_column,split_value)
    
        # check for emply data in a particular split
        if len(data_below) == 0 or len(data_above) == 0:
            classification = classify(data)
            return classification
        
        # subtree
        feature_name = COLUMN_HEADERS[split_column]
        type_of_feature = FEATURE_TYPES[split_column] 
        
        if type_of_feature == "continuous":
            question = "{} <= {}".format(feature_name,split_value)
        else: # if the feature is categorical
            question = "{} = {}".format(feature_name,split_value)
        Sub_tree = {question: []}
    
    # find answers 
    yes_answer =  decision_tree_algorithm(data_below,counter,min_samples,max_depth)
    no_answer =  decision_tree_algorithm(data_above,counter,min_samples,max_depth)
    
    if yes_answer == no_answer:
        Sub_tree = yes_answer
    else:       
        Sub_tree[question].append(yes_answer)
        Sub_tree[question].append(no_answer)
    
    return Sub_tree
```

```python
train_df,test_df = train_test_split(df, test_size=.10)
tree = decision_tree_algorithm(train_df,max_depth=5)
accuracy = calculate_accuracy(test_df,tree)
pprint(tree,width = 50)
accuracy
```
